```python
import numpy as np

def hosvd(tensor):
    # Initialize list to store the U matrices
    U_matrices = []
    # Loop over each mode of the tensor
    for mode in range(tensor.ndim):
        # Unfold the tensor along this mode
        unfold = np.moveaxis(tensor, mode, 0)
        unfold = unfold.reshape(unfold.shape[0], -1)
        
        # Compute the SVD of the unfolded tensor
        U, S, Vt = np.linalg.svd(unfold, full_matrices=False)
        # Store the U matrix
        U_matrices.append(U)
    
    # Compute the core tensor S by multiplying the tensor with the U matrices along each mode
    S_tensor = tensor 
    for mode, U in enumerate(U_matrices):
        S_tensor = np.tensordot(U.T, S_tensor, axes=([1],[mode]))
        S_tensor = np.moveaxis(S_tensor, 0, mode)  # move the transformed axis back to its original position
    
    return U_matrices, S_tensor


# Example usage:
T = np.arange(0, 24).reshape((3,4,2))
tensor = np.array([[[ 0.9073,  0.7158,  -0.3698],
                    [ 0.8924,  -0.4898,  2.4288],
                    [ 2.1488,  0.3054,  2.3753]],
                   
                   [[1.7842, 1.6970, 0.0151],
                    [1.7753, -1.5077, 4.0337],
                    [4.2495, 0.3207, 4.7146]],
                   
                   [[2.1236, -0.0740, 1.4429],
                    [-0.6631, 1.9103, -1.7495],
                    [1.8260, 2.1335, -0.2716]]])

# tensor = np.random.rand(5, 4, 3)  # Random 3rd order tensor
U_matrices, S_tensor = hosvd(tensor)

for i, U_matrix in enumerate(U_matrices):
    print(f"\nFactor Matrix {i}:")
    print(U_matrix)
#print(U_matrices)
print(S_tensor)
```

```python
import numpy as np

def tensor_train_classification(z, G1, G2, G3):
    # Compute z' = G1^T*z
    z_prime = np.dot(G1.T, z)
    
    # Initialize the DTT matrix
    n_e, n_p = G2.shape[0], G3.shape[1]
    DTT = np.zeros((n_e, n_p))

    # For each mode of G2
    for e in range(n_e):
        # Contract G2[e,:,:] and z' to solve for _e
        _e = np.tensordot(G2[e,:,:], z_prime, axes=1)

        # For each column of G3
        for p in range(n_p):
            # Compute DTT(e, p) = distance(e  g3^p)
            g3_p = G3[:,p]
            DTT[e,p] = np.linalg.norm(_e - g3_p)
    
    # Find e', p' that minimize DTT
    e_prime, p_prime = np.unravel_index(np.argmin(DTT, axis=None), DTT.shape)

    # Return the person classification
    return p_prime
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorly as tl
from tensorly.decomposition import matrix_product_state
from PIL import Image

# Open the image file
img = Image.open("bear.jpg")
# Convert the image to a numpy array
image_array = np.array(img)

def compress_image(image_array, rank):
    """Compress an image array using Tensor Train decomposition."""
    # Rescale image values to [0, 1]
    image_tensor = image_array / 255
    # Decompose the image tensor
    factors = matrix_product_state(image_tensor, rank)
    return factors

def reconstruct_image(factors):
    """Reconstruct an image array from a Tensor Train decomposition."""
    # Reconstruct the tensor from the decomposed factors
    reconstructed_tensor = tl.tt_to_tensor(factors)
    # Rescale the tensor values to [0, 255]
    reconstructed_array = np.clip(reconstructed_tensor * 255, 0, 255).astype(np.uint8)
    return reconstructed_array

# Compress the image tensor using Tensor Train
rank = [1, 10, 10, 1]
factors = compress_image(image_array, rank)

# Reconstruct the image
reconstructed_image = reconstruct_image(factors)

# Display the original and reconstructed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_array)
plt.title("Original")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image)
plt.title("Reconstructed")
plt.axis("off")
plt.show()
```

```python
import numpy as np
from tensorly.decomposition import tucker
import tensorly as tl
from keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
def base_matrices(dig):
    # Select 100 images of digit "dig"
    digits = train_images[train_labels == dig][:1000]
    digits = np.transpose(digits, (1, 2, 0))

    # Reshape the 3D tensor
    A = digits.reshape(28, 28, -1)

    # Perform HOSVD
    core, factors = tucker(A, rank=[28, 28, 10])  # Adjust the rank to match number of base matrices

    # Get the factor matrices
    u1, u2, u3 = factors
    
    # Base matrices
    As = []
    for i in range(10):  # Range adjusted to 10
        # Perform tensor-tensor product to get base matrix
        S = np.zeros((28, 28, 10))  # Adjusted third dimension to 10
        S[:, :, i] = core[:, :, i]
        A = tl.tucker_to_tensor((S, [u1, u2, u3]))
        # Convert to 2D matrix by selecting the i-th matrix along the third dimension
        A = A[:, :, i]
        As.append(A)
    return As

print(len(base_matrices(4)))
```

```python
import numpy as np

def calculate_error(Z, A_list):
    assert len(A_list) == 10, "List should contain 10 basis tensors"

    # Unpacking the list
    A1, A2, A3, A4, A5, A6, A7, A8, A9, A10 = A_list

    # Calculating coefficients for the basis
    z1 = np.sum(Z * A1) / np.sum(A1 * A1)
    z2 = np.sum(Z * A2) / np.sum(A2 * A2)
    z3 = np.sum(Z * A3) / np.sum(A3 * A3)
    z4 = np.sum(Z * A4) / np.sum(A4 * A4)
    z5 = np.sum(Z * A5) / np.sum(A5 * A5)
    z6 = np.sum(Z * A6) / np.sum(A6 * A6)
    z7 = np.sum(Z * A7) / np.sum(A7 * A7)
    z8 = np.sum(Z * A8) / np.sum(A8 * A8)
    z9 = np.sum(Z * A9) / np.sum(A9 * A9)
    z10 = np.sum(Z * A10) / np.sum(A10 * A10)

    # Calculating total error
    error = Z - z1 * A1 - z2 * A2 - z3 * A3 - z4 * A4 - z5 * A5 - z6 * A6 - z7 * A7 - z8 * A8 - z9 * A9 - z10 * A10
    error = np.sqrt(np.sum(error * error))

    return error
```

```python
def accuracy():
    # Base matrices for each digit
    base_matrices_list = [base_matrices(dig) for dig in range(10)]

    # Traverse the test set
    n = len(test_labels)
    est = np.zeros((n, 1))
    error = [0] * 10
    
    for i in range(n):
        # Image 'Z' from the test set
        Z = test_images[i]

        for j in range(10):
            error[j] = calculate_error(Z, base_matrices_list[j])

        # Classification of the image 'Z'
        est[i, 0] = np.argmin(error)

    # Overall accuracy of the algorithm
    res = np.column_stack((test_labels, est))
    accuracy = np.sum(res[:, 0] == res[:, 1]) / n * 100
    accuracy_message = f"Accuracy is {accuracy}%."
    print(accuracy_message)
```

```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorly.decomposition import matrix_product_state
from tensorly import tt_to_tensor

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# We will use 100 images per digit for the bases
bases = []

for digit in range(10):
    images = train_images[train_labels == digit][:100]
    # Reshape images to the correct shape and perform TT decomposition
    images_reshaped = images.reshape((100, 28, 28))
    # Generate tensor trains for each image and average them to create a single base for each digit
    avg_tt = np.mean([matrix_product_state(image.reshape((1, 28, 28)), rank=[1, 20, 20, 1]) for image in images_reshaped], axis=0)
    bases.append(avg_tt)

# We will classify 1000 test images
test_images = test_images[:1000]
test_labels = test_labels[:1000]

predictions = []

for image in test_images:
    image_reshaped = image.reshape((1, 28, 28))
    tt_image = matrix_product_state(image_reshaped, rank=[1, 20, 20, 1])

    # Classify based on least approximation error with the bases
    errors = [np.mean((tt_to_tensor(base) - tt_to_tensor(tt_image)) ** 2) for base in bases]
    prediction = np.argmin(errors)
    predictions.append(prediction)

# Calculate the accuracy
accuracy = np.mean(np.array(predictions) == test_labels)

# Express accuracy as a percentage
accuracy_percent = accuracy * 100
print('Accuracy:', accuracy_percent, "%")
```
